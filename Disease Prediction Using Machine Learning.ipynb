```python
# Identify the independent and Target (dependent) variables

IndepVar = []
for col in train.columns:
    if col != 'prognosis':
        IndepVar.append(col)

TargetVar = 'prognosis'

x = train[IndepVar]
y = train[TargetVar]
```

```python
# Split the data into train and test

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.7,random_state = 42)

# display the shape fortrain & test data

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf= RandomForestClassifier(random_state=42)
model_rf = rf.fit(xtrain,ytrain)
tr_pred_rf = model_rf.predict(xtrain)
ts_pred_rf = model_rf.predict(xtest)

print("training accuracy is:",accuracy_score(ytrain,tr_pred_rf)*100)
print("testing accuracy is:",accuracy_score(ytest,ts_pred_rf)*100)
```
